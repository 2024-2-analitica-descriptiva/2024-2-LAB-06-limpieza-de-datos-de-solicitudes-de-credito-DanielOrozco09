```python
import pandas as pd

# Cargar el archivo CSV
df = pd.read_csv(r'D:\Unal\Analitica Descriptiva\Labs\2024-2-LAB-06-limpieza-de-datos-de-solicitudes-de-credito-DanielOrozco09\files\input\solicitudes_de_credito.csv', sep=';')

# Cambiar todos los datos tipo string de mayúscula a minúscula
df = df.applymap(lambda s: s.lower() if type(s) == str else s)

# Eliminar duplicados
df.drop_duplicates(subset= ['Unnamed: 0'], inplace=True)

# Darle el mismo formato a todas las fechas
df['fecha_de_beneficio'] = pd.to_datetime(df['fecha_de_beneficio'], errors='coerce', dayfirst=True)
df['fecha_de_beneficio'] = df['fecha_de_beneficio'].dt.strftime('%Y-%m-%d')

# Eliminar registros con falta de datos
df.dropna(inplace=True)

# Guardar el DataFrame limpio en un nuevo archivo CSV
df.to_csv('../files/output/solicitudes_de_credito.csv', index=False, sep=';')


df
```
